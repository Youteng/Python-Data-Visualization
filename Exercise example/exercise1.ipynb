```python
import requests
from bs4 import BeautifulSoup
```

```python
#所要擷取的網站網址
url = "https://www.ptt.cc/bbs/Gossiping/M.1630476638.A.0A9.html"

#設定Header與Cookie
headers = {'cookie': 'over18=1;'}
#發送get 請求到 ptt 
response = requests.get(url, headers = headers)
#將擷取的網站印出
print(response.text)
```

```python
#將原始碼做整理
soup = BeautifulSoup(response.text, 'lxml')
#使用find_all()找推文
pushes = soup.find_all('div', 'push')
#印出推文
print(pushes)
```

```python
#取得推文內容
for push in pushes:
    message = push.find('span', 'f3 push-content').getText()
    print(message)
```

```python
#將推文寫入檔案中
with open('exercise1.txt','w', encoding='utf-8') as fp:
    for push in pushes:
    
        #去除掉冒號和左右的空白
        message = push.find('span','f3 push-content').getText().replace(':','').strip()
        print(message)
        fp.write(message + "\n")
```

```python
#導入函式庫
%matplotlib inline
import wordcloud
import matplotlib.pyplot as plt
import numpy as np
from janome.tokenizer import Tokenizer
from PIL import Image
import pandas as pd
```

```python
text_ptt = open('exercise1.txt','r',  encoding='utf-8').read()

#利用wordcloud函式庫將文字資訊畫成文字雲
wc = wordcloud.WordCloud(width=1000, height=600, background_color="white", 
                         font_path=r"C:\Windows\Fonts\msjh.ttc")  #字型的檔案路徑


wc.generate(text_ptt)
plt.imshow(wc)
plt.axis("off")
```

```python
#載入遮罩圖片
mask_image = np.array(Image.open("circle.png"))

wc = wordcloud.WordCloud(width=700,
                         height=700,
                         background_color="white",
                         font_path=r"C:\Windows\Fonts\msjh.ttc",
                         mask=mask_image,
                         contour_width=6,    #邊框的粗細
                         contour_color="pink",    #邊框的顏色
                         colormap="plasma"    #文字雲的調色盤
                         )

wc.generate(text_ptt)
plt.imshow(wc)
plt.axis("off")
```
